```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mu_stick = 0.33
sigma_stick = 0.1
mu_switch = 0.67
sigma_switch = 0.1

alpha = 0.05

x = np.linspace(0, 1, 1000)

pdf_stick = norm.pdf(x, mu_stick, sigma_stick)
pdf_switch = norm.pdf(x, mu_switch, sigma_switch)

z_alpha = norm.ppf(1 - alpha, mu_stick, sigma_stick)

x_alpha = np.linspace(z_alpha, 1, 100)
y_alpha = norm.pdf(x_alpha, mu_stick, sigma_stick)

x_beta = np.linspace(0, z_alpha, 100)
y_beta = norm.pdf(x_beta, mu_switch, sigma_switch)

plt.figure(figsize=(12, 8))
plt.plot(x, pdf_stick, label='Distribui√ß√£o H0 (Stick)', color='blue')
plt.plot(x, pdf_switch, label='Distribui√ß√£o H1 (Switch)', color='green')

plt.fill_between(x_alpha, y_alpha, color='blue', alpha=0.3, label='Œ± (Erro tipo I)')

plt.fill_between(x, pdf_stick, where=(x < z_alpha), color='blue', alpha=0.1, label='1-Œ± (N√≠vel de confian√ßa)')

plt.fill_between(x_beta, y_beta, color='green', alpha=0.3, label='Œ≤ (Erro tipo II)')

plt.fill_between(x, pdf_switch, where=(x >= z_alpha), color='green', alpha=0.1, label='1-Œ≤ (Poder do teste)')

plt.axvline(z_alpha, color='red', linestyle='dashed', linewidth=1, label='Ponto de corte para Œ±')

plt.title('Distribui√ß√µes das Estrat√©gias Stick e Switch no Problema de Monty Hall')
plt.xlabel('Probabilidade de Ganhar')
plt.ylabel('Densidade')
plt.legend(loc='upper left')
plt.grid(True)

plt.show()
```
