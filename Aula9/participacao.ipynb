```python
#sampler
def shovel(bowl,number_estimatives, number_balls):
    values = np.empty(number_estimatives)
    for i in range(number_estimatives):
        sample = bowl.sample(n=number_balls)
        sample = sample[sample[0] == "red"]
        values[i] = sample.value_counts().iloc[0]/number_balls

    return values

shovel(bowl,110,100)
```

```python
def bootstrap_confidence_interval(data, number_samples=1000, confidence_level=0.95):
    sample_means = np.empty(number_samples)
    n = len(data)
    for i in range(number_samples):
        sample = np.random.choice(data, size=n, replace=True)
        sample_means[i] = np.mean(sample)
    return sample_means
```

    R: Método percentil:

```python
confidence_level = 0.95
sample_means = bootstrap_confidence_interval(pennies_sample['year'].values)
lower_bound = np.percentile(sample_means, ((1 - confidence_level) / 2) * 100)
upper_bound = np.percentile(sample_means, (1 - (1 - confidence_level) / 2) * 100)
f"Intervalo de confiança (percentil): {lower_bound} , {upper_bound}"
```

#### 6. Visualize, com base nas distribuições das estratégias stick e switch, as quatro probabilidades: α (taxa de erro tipo I), (1 − α) (níveis de confiança), β (taxa de erro tipo II) e (1 − β) (poder do teste) para a hipótese de que “é mais benéfico manter a porta inicialmente escolhida” sob um nível de significância de 0.05 no contexto do problema de Monty Hall, discutido no item 4 da Seção 7.8.

```python
def mounty_hall(estrategia):
    doors = np.arange(1,4)
    prize = np.array(("car","goat","goat"))
    prize_door = doors[prize == "car"]
    my_pick = np.random.choice(doors,1)
    show = np.random.choice(doors[~np.in1d(doors, [my_pick[0],prize_door[0]])],1)
    stick = my_pick
    stick == prize_door
    switch = doors[~np.in1d(doors, [my_pick[0], show[0]])]
    if estrategia == "stick":
        choice = stick
    else:
        choice = switch
    return choice == prize_door

def simulate_strategy(estrategia, num_simulations):
    count_true = 0
    for _ in range(num_simulations):
        if mounty_hall(estrategia):
            count_true += 1
    return count_true/ num_simulations

# mouty hall
stick = []
switch = []
for i in range(1000):
    stick.append(simulate_strategy("stick", 30))
    switch.append(simulate_strategy("switch", 30))
```

```python
stick_df = pd.DataFrame(stick)
stick_df["strategy"] = np.repeat("stick", len(stick))
stick_df

switch_df = pd.DataFrame(switch)
switch_df["strategy"] = np.repeat("switch", len(switch))
switch_df

concat = pd.concat([stick_df,switch_df])
concat
```

```python
seed = mythbusters[mythbusters["group"] == "seed"]
seed = seed["yawn"]
control = mythbusters[mythbusters["group"] == "control"]
control = control["yawn"]
```

```python
seed.value_counts()["yes"]/len(seed)
```

```python
def bootstrap_confidence_interval(data, number_samples=1000, confidence_level=0.95):
    sample_means = np.empty(number_samples)
    n = len(data)
    for i in range(number_samples):
        sample = np.random.choice(data, size=50, replace=True)
        _,x = np.unique(sample, return_counts=True, return_index=False)
        sample_means[i] = x[1]/len(sample)
    return sample_means
bootstrap_confidence_interval(seed).mean()
```

```python
confidence_level = 0.95
sample_means = bootstrap_confidence_interval(control)
lower_bound = np.percentile(sample_means, ((1 - confidence_level) / 2) * 100)
upper_bound = np.percentile(sample_means, (1 - (1 - confidence_level) / 2) * 100)
f"Intervalo de confiança de quem não viu bocejo: {lower_bound} , {upper_bound}"
```

Na Seção 9.1 em [38], os autores formularam a pergunta como duas hipóteses 𝐻0 e 𝐻1 e aplicaram a técnica de bootstrap para respondê-la. Plote o gráfico de distribuição das duas hipóteses, hachurando a área correspondente ao p-valor.

    Com H0 sendo que não existe diferença e H1 que homens são mais promovidos:

```python
import pandas as pd
import numpy as np
from plotnine import *
```

```python
promotions = pd.read_csv("promotions.csv")
promotions.head()
```

```python
#visualize
ggplot(promotions) \
+ aes(x="gender", fill = "decision") \
+ geom_bar()
```

```python
valores = promotions[["gender","decision"]].value_counts()
p_male = valores["male"]["promoted"]/valores["male"].sum()
p_female = valores["female"]["promoted"]/valores["female"].sum()
p_value = p_male - p_female
p_value
```

```python
shuffled_df = promotions.copy()
shuffled_df["gender"] = shuffled_df["gender"].sample(frac=1).values
ggplot(shuffled_df) \
+ aes(x="gender", fill = "decision") \
+ geom_bar()
```

```python
#shuffle
def shuffle(df,shuffle_number=1000):
    h0 = []

    for i in range(shuffle_number):
        shuffled_df = df.copy()
        shuffled_df["gender"] = shuffled_df["gender"].sample(frac=1).values
        shuffled_df_valores = shuffled_df[["gender","decision"]].value_counts()
        p_male = shuffled_df_valores["male"]["promoted"]/shuffled_df_valores["male"].sum()
        p_female = shuffled_df_valores["female"]["promoted"]/shuffled_df_valores["female"].sum()
        h0.append(p_male - p_female)
    return h0

shuffle(promotions,10)[:]
```

```python
h0 = pd.DataFrame(shuffle(promotions,1000))
h0[0]
ggplot() + \
aes(h0[0]) + \
geom_histogram(binwidth=0.1,color="black", fill="green") + \
geom_vline(xintercept=p_value)
```
