```python
#criar algo como dataset bowl 
bowl_list = [] #2400
for i in range(900):
    bowl_list.append("red")
for i in range(1500):
    bowl_list.append("white")
bowl = pd.DataFrame(bowl_list)
bowl = bowl.sample(n=2400)
bowl = pd.DataFrame(bowl.values)
bowl
```

```python
#sampler
def shovel(bowl,number_estimatives, number_balls):
    values = np.empty(number_estimatives)
    for i in range(number_estimatives):
        sample = bowl.sample(n=number_balls)
        sample = sample[sample[0] == "red"]
        values[i] = sample.value_counts().iloc[0]/number_balls

    return values

shovel(bowl,110,100)
```

```python
case1 = pd.DataFrame(shovel(bowl,100,400))
case2 = pd.DataFrame(shovel(bowl,400,400))
case3 = pd.DataFrame(shovel(bowl,100,100))
case4 = pd.DataFrame(shovel(bowl,400,100))
case1["origem"] = "case1"
case2["origem"] = "case2"
case3["origem"] = "case3"
case4["origem"] = "case4"
all = pd.concat([case1, case2, case3, case4], axis=0)
all.columns = ["valor","origem"]
all.reset_index(drop=True,inplace=True)
# all
```

```python
desc = pennies_sample['year'].describe()
print(desc)
```

```python
def bootstrap_confidence_interval(data, number_samples=1000, confidence_level=0.95):
    sample_means = np.empty(number_samples)
    n = len(data)
    for i in range(number_samples):
        sample = np.random.choice(data, size=n, replace=True)
        sample_means[i] = np.mean(sample)
    return sample_means
```

    R: Método percentil:

```python
confidence_level = 0.95
sample_means = bootstrap_confidence_interval(pennies_sample['year'].values)
lower_bound = np.percentile(sample_means, ((1 - confidence_level) / 2) * 100)
upper_bound = np.percentile(sample_means, (1 - (1 - confidence_level) / 2) * 100)
f"Intervalo de confiança (percentil): {lower_bound} , {upper_bound}"
```

    R: Método de Erro Padrão

```python
sample_means = bootstrap_confidence_interval(pennies_sample['year'].values)
xbar = sample_means.mean()
SE = np.std(sample_means)
lower_bound = xbar - 1.96*SE
upper_bound = xbar + 1.96*SE

f"Intervalo de confiança (percentil): {lower_bound} , {upper_bound}"
```

```python
def bootstrap_confidence_interval(data, number_samples=1000, confidence_level=0.95):
    sample_means = np.empty(number_samples)
    n = len(data)
    for i in range(number_samples):
        sample = np.random.choice(data, size=50, replace=True)
        _,x = np.unique(sample, return_counts=True, return_index=False)
        sample_means[i] = x[1]/len(sample)
    return sample_means
bootstrap_confidence_interval(seed).mean()
```
